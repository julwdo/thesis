```python
import os, json, ijson, decimal
from google.colab import drive

drive.mount('/content/drive')

in_dir = "/content/drive/MyDrive/TwiBot-22"

out_dir = "/content/drive/MyDrive/twibot-22/raw"
os.makedirs(out_dir, exist_ok=True)

files_to_process = ["tweet_0.json", "user.json"]

def json_default(o):
    if isinstance(o, decimal.Decimal):
        if o % 1 == 0:
            return int(o)
        return float(o)
    raise TypeError(f"Object of type {o.__class__.__name__} is not JSON serializable")

for fname in files_to_process:
    in_path = os.path.join(in_dir, fname)
    out_fname = fname.replace(".json", ".jsonl")
    out_path = os.path.join(out_dir, out_fname)

    print(f"Processing {in_path} -> {out_path}")

    with open(in_path, "r") as f, open(out_path, "w") as out:
        parser = ijson.items(f, "item")
        for i, obj in enumerate(parser, start=1):
            out.write(json.dumps(obj, default=json_default) + "\n")
            if i % 10000 == 0:
                print(f"  {i} objects written locally...")

    print(f"Finished {fname}, saved as {out_path}")

print("All selected files processed and saved to Google Drive.")
```
